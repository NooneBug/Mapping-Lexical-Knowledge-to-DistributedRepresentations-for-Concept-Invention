## Retrieve all URI and preprocess them

```python
sons_of = {}

sons_of['Activity'] = ['Sport']
sons_of['Agent'] = ['FictionalCharacter']
sons_of['Organisation'] = ['Company', 
                           'SportsTeam']
sons_of['Thing'] = ['Award', 
                    'Colour',
                    'EthnicGroup', 
                    'Language',
                    'Plant', 
                    'Weapon']
sons_of['Place'] = ['ArchitecturalStructure', 
                    'Planet']
sons_of['Animal'] = ['Bird',
                     'Insect', 
                     'Mammal']
# sons_of['Person'] = ['Artist', 'Scientist'
#                     ,'Athlete', 'Writer'
#                     ]
```

```python
types = ['dbo:' + s for sublist in sons_of.values() for s in sublist]

classes = [x.replace('dbo:', '') for x in types]
```

### retrieve entities which have a composite name (in this case word phrase with length = 2) like 'ice hokey' from dbr:Ice_Hokey

```python
words = []
import re
pat = re.compile('[A-Z][a-z]*[A-Z][a-z]*')
midpat = re.compile('[A-Z][a-z]*')
for i, t in enumerate(types):
    w = get_from_class(t)
    if pat.findall(t):
        to_append = midpat.findall(t)[0] + '_' + midpat.findall(t)[1]
    else:
        to_append = t.replace('dbo:', '') 
    w.append(to_append)
    words.append(w) 
    clear_output()
    print('{}%'.format(int(round((i+1)/len(types), 2) * 100)))
```

### some entities has something between parenthesis or are divided with '-'
#### only word phrases with length = WORD_PHRASE_LENGTH are mantained

```python
comp_words = {c: [] for c in classes}
regex = re.compile(".*?\((.*?)\)")

WORD_PHRASE_LENGTH = 2

for i, w_list in enumerate(words):
    for w in w_list:
        w = re.sub(r'\([^)]*\)', '', w).strip()
        w = w.replace("'", "")
        
        if len(w.split('_')) == WORD_PHRASE_LENGTH and len(w.split('-')) < WORD_PHRASE_LENGTH:
            if '' not in w.lower().split('_'):
                comp_words[classes[i]].append(w.lower().split('_'))
        elif len(w.split('-')) == WORD_PHRASE_LENGTH and len(w.split('_')) < WORD_PHRASE_LENGTH:
            if '' not in w.lower().split('-'):
                comp_words[classes[i]].append(w.lower().split('-'))    
```

### Ariety of found word phrases

```python
alpha = re.compile('^[A-z]*$')
word_to_index = set([s for sublists in comp_words.values() for sublist in sublists for s in sublist if re.search(alpha, s) and s != ''])
len(word_to_index)
```

### retrieve in which sentence a single word appears

```python
word_index = {}

prev = 0
length = len(liness)
for i, line in enumerate(liness):
    for word in line:
        if word in word_to_index:
            try:
                word_index[word].append(i)
            except:
                word_index[word] = [i]
    
    perc = int(round(i / length, 2)*100)
    
    if prev != perc:
        clear_output()
        print('{}%'.format(perc))
        prev = perc
        
word_index = {k:set(v) for k,v in word_index.items()}
```

#### for each word phrase retrieve in which sentence all words appear

```python
comp_word_indexes = {c: [] for c in classes}
for k, typelist in comp_words.items():
    for couple in typelist:
        try:
            inter = word_index[couple[0]].intersection(word_index[couple[1]])
            if inter:
                comp_word_indexes[k].append([couple, list(inter)])
        except:
            pass
```

```python
{k: len(v) for k, v in comp_word_indexes.items()}
```

#### retrieve index of word phrases based on the position of words in the sentence (which has to be contiguous, see this in the 'if')

```python
real_couples_indexes = {}
tot = len(comp_word_indexes.values())
j = 0
for j, (k, v) in enumerate(comp_word_indexes.items()):
    for couple in v:
        for index in couple[1]:
            if abs(liness[index].index(couple[0][0]) - 
                   liness[index].index(couple[0][1])) == 1:
                try:
                    real_couples_indexes["{}_{}".format(couple[0][0], couple[0][1])].append([
                        [liness[index].index(couple[0][0]), liness[index].index(couple[0][1])],
                        index])
                except:
                    real_couples_indexes["{}_{}".format(couple[0][0], couple[0][1])] = [k, 
                                                                                        [
                                                                                            [liness[index].index(couple[0][0]), 
                                                                                             liness[index].index(couple[0][1])],
                                                                                            index]]
    clear_output()
    print('{}%'.format(int(round((j + 1)/tot, 2)) * 100))
```
